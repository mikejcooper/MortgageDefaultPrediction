```python
df = df.sort_values("svcg_cycle", ascending=False)
index_last_loan_entries = df['id_loan'].drop_duplicates().index
df_last_loan_entries = df.loc[index_last_loan_entries]
df1 = df_last_loan_entries

df1 = df1.sort_values("svcg_cycle", ascending=True)

columns = ['svcg_cycle', 'zipcode', 'label_good_bad_loan']

local_data_zipcode = pd.DataFrame() 
local_data_zipcode_loan_count = pd.DataFrame() 
local_data_zipcode_paid_loans = pd.DataFrame() 

current_data = df1[columns].drop_duplicates()



data_source = [current_data, current_data]    
for new_data in data_source:
    
    new_data.loc[:, 'loan_count'] = 1
    new_data_loan_count = new_data[['svcg_cycle', 'zipcode', 'loan_count']]
    new_data_paid_count = new_data[['svcg_cycle', 'zipcode', 'label_good_bad_loan']]
    
    
    
#     new_data_loan_count.loc[0] = [1,1,5]
#     new_data_loan_count.loc[1] = [1,1,2]
#     new_data_loan_count.loc[2] = [1,2,11]
    
#     new_data_paid_count.loc[0] = [1,1,5]
#     new_data_paid_count.loc[1] = [1,1,2]
#     new_data_paid_count.loc[2] = [1,2,10]
    
# #     new_data_paid_count.loc[0] = [1,2,5]
# #     new_data_paid_count.loc[1] = [1,2,11]
    
    new_data_loan_count['loan_count'] = new_data_loan_count.groupby(['zipcode', 'svcg_cycle'])['loan_count'].transform('sum')
    new_data_loan_count = new_data_loan_count.drop_duplicates()
    
    new_data_paid_count['label_good_bad_loan'] = new_data_paid_count.groupby(['zipcode', 'svcg_cycle'])['label_good_bad_loan'].transform('sum')
    new_data_paid_count = new_data_paid_count.drop_duplicates()
    
    if(local_data_zipcode_loan_count.shape[0] < 10):
        local_data_zipcode_loan_count = new_data_loan_count
        local_data_zipcode_paid_loans = new_data_paid_count
        continue
    
    
    local_data_zipcode_loan_count = pd.merge(local_data_zipcode_loan_count, new_data_loan_count, on=['svcg_cycle', 'zipcode']).set_index(['svcg_cycle', 'zipcode']).sum(axis=1)
    local_data_zipcode_loan_count = local_data_zipcode_loan_count.to_frame()
    local_data_zipcode_loan_count.reset_index(inplace=True)
    local_data_zipcode_loan_count = local_data_zipcode_loan_count.rename(columns = {0:'loan_count'})
    local_data_zipcode_loan_count['loan_count'] = local_data_zipcode_loan_count.groupby(['zipcode', 'svcg_cycle'])['loan_count'].transform('sum')
    local_data_zipcode_loan_count = local_data_zipcode_loan_count.drop_duplicates()
 
    local_data_zipcode_paid_loans = pd.merge(local_data_zipcode_paid_loans, new_data_paid_count, on=['svcg_cycle', 'zipcode']).set_index(['svcg_cycle', 'zipcode']).sum(axis=1)
    local_data_zipcode_paid_loans = local_data_zipcode_paid_loans.to_frame()
    local_data_zipcode_paid_loans.reset_index(inplace=True)
    local_data_zipcode_paid_loans = local_data_zipcode_paid_loans.rename(columns = {0:'paid_loans'})
    local_data_zipcode_paid_loans['paid_loans'] = local_data_zipcode_paid_loans.groupby(['zipcode', 'svcg_cycle'])['paid_loans'].transform('sum')
    local_data_zipcode_paid_loans = local_data_zipcode_paid_loans.drop_duplicates()
    
    local_data_zipcode = pd.merge(local_data_zipcode_paid_loans, local_data_zipcode_loan_count, on=['svcg_cycle', 'zipcode']).set_index(['svcg_cycle', 'zipcode'])
    local_data_zipcode.reset_index(inplace=True)

    local_data_zipcode = local_data_zipcode.sort_values("svcg_cycle", ascending=True)
    
    local_data_zipcode_paid_loans = local_data_zipcode_paid_loans.sort_values(['zipcode', 'svcg_cycle'], ascending=[True, True])
    
    # print df3.loc[df3['svcg_cycle'] == 200112]
    
#     print local_data_zipcode.loc[local_data_zipcode['paid_loans'] != 2]

df1 = pd.merge(df, local_data_zipcode, on=['svcg_cycle', 'zipcode']).set_index(['svcg_cycle', 'zipcode'])
df1.reset_index(inplace=True)
df1 = df1.sort_values("id_loan", ascending=True)

print df1[['id_loan','svcg_cycle', 'zipcode', 'paid_loans']]

```
