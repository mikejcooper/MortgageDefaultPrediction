```python
import random
import datetime

svcg_cycles = [   
    199901, 199902, 199903, 199904, 199905, 199906, 199907, 199908, 199909, 199910, 199911, 199912,
    200001, 200002, 200003, 200004, 200005, 200006, 200007, 200008, 200009, 200010, 200011, 200012, 
    200101, 200102, 200103, 200104, 200105, 200106, 200107, 200108, 200109, 200110, 200111, 200112, 
    200201, 200202, 200203, 200204, 200205, 200206, 200207, 200208, 200209, 200210, 200211, 200212, 
    200301, 200302, 200303, 200304, 200305, 200306, 200307, 200308, 200309, 200310, 200311, 200312, 
    200401, 200402, 200403, 200404, 200405, 200406, 200407, 200408, 200409, 200410, 200411, 200412,
    200501, 200502, 200503, 200504, 200505, 200506, 200507, 200508, 200509, 200510, 200511, 200512, 
    200601, 200602, 200603, 200604, 200605, 200606, 200607, 200608, 200609, 200610, 200611, 200612, 
    200701, 200702, 200703, 200704, 200705, 200706, 200707, 200708, 200709, 200710, 200711, 200712, 
    200801, 200802, 200803, 200804, 200805, 200806, 200807, 200808, 200809, 200810, 200811, 200812, 
    200901, 200902, 200903, 200904, 200905, 200906, 200907, 200908, 200909, 200910, 200911, 200912, 
    201001, 201002, 201003, 201004, 201005, 201006, 201007, 201008, 201009, 201010, 201011, 201012,
    201101, 201102, 201103, 201104, 201105, 201106, 201107, 201108, 201109, 201110, 201111, 201112, 
    201201, 201202, 201203, 201204, 201205, 201206, 201207, 201208, 201209, 201210, 201211, 201212, 
    201301, 201302, 201303, 201304, 201305, 201306, 201307, 201308, 201309, 201310, 201311, 201312, 
    201401, 201402, 201403, 201404, 201405, 201406, 201407, 201408, 201409, 201410, 201411, 201412, 
    201501, 201502, 201503, 201504, 201505, 201506, 201507, 201508, 201509, 201510, 201511, 201512,
    201601, 201602, 201603, 201604, 201605, 201606, 201607, 201608, 201609, 201610, 201611, 201612, 
    201701, 201702, 201703, 201704,
]

def get_N_random_loan_ids(N, df):
    df_1 = df.drop_duplicates(subset='id_loan')
    set_size = N
    random.seed(len(df_1) + datetime.datetime.now().second + 1)
    sudo_ran_sample = [random.randint(1, len(df_1)) for i in range(1, len(df_1))]
    msk = np.array(sudo_ran_sample) < N + 20
    ids_N = df.iloc[msk]['id_loan']
    return ids_N[:N].values

def fill_dataframe_to_svcg_cycle(df_IN):
    df_svcg_cycle = pd.DataFrame(svcg_cycles, columns=['svcg_cycle'])
    df_filled = pd.merge(df_svcg_cycle, df_IN, on='svcg_cycle', how='outer')
    df_filled['current_upb_ratio'] = df_filled['current_upb_ratio'].transform(lambda v: v.ffill())
    df_filled['current_upb_ratio'] = df_filled['current_upb_ratio'].fillna(0)
    return df_filled

def get_loan_data(loan_id_IN, df_IN):
    return df_IN.loc[df_IN['id_loan'] == loan_id_IN]
    
def get_payment_history(loan_data_IN):
    loan_data_IN = loan_data_IN.sort_values(['svcg_cycle'], ascending=[True])
    # create ratio column 
    loan_amount = loan_data_IN['current_upb'].values[0]
    loan_data_IN['current_upb_ratio'] = loan_data_IN['current_upb'] / loan_amount    
    # Drop last if bad loan      
    label_good_bad_loan = loan_data_IN['label_good_bad_loan'].values[0]
    if label_good_bad_loan == 0:
        loan_data_IN = loan_data_IN.drop(loan_data_IN.index[len(loan_data_IN)-1])
    return loan_data_IN[['svcg_cycle', 'current_upb_ratio']]
    
def get_combined_payment_history(payment_history_IN):
    df_combined = payment_history_IN[0]
    df_combined = df_combined.drop('current_upb_ratio', 1)
    for payment_history_i in payment_history_IN:
        payment_history_i['ones'] = 1
        df_combined = pd.merge(payment_history_i, df_combined, on='svcg_cycle', how='outer')
        df_combined = df_combined.drop('current_upb_ratio', 1)
        df_combined = df_combined.drop('ones', 1)
    
    for payment_history_i in payment_history_IN:
        df_combined = pd.merge(payment_history_i, df_combined, on='svcg_cycle', how='outer')
        
    current_upb_cols = []
    ones_cols = []
    for col in df_combined.columns.values:
        if "current_upb_ratio_" in col:
            current_upb_cols.append(col)
            df_combined[col] = df_combined[col].transform(lambda v: v.ffill())
        if "ones_" in col:
            ones_cols.append(col)
            df_combined[col] = df_combined[col].transform(lambda v: v.ffill())
    
    df_combined['current_upb_ratio'] = df_combined[current_upb_cols].sum(axis=1)
    df_combined['ones'] = df_combined[ones_cols].sum(axis=1)
    df_combined['current_upb_ratio'] = 1 - df_combined['current_upb_ratio'] / df_combined['ones']
    df_combined = df_combined.drop('ones', 1)
    
    
    current_upb_cols = np.unique(current_upb_cols)
    ones_cols = np.unique(ones_cols)
    
    for col in current_upb_cols:
        df_combined = df_combined.drop(col, 1)
    
    for col in ones_cols:
        df_combined = df_combined.drop(col, 1)
            
    df_combined = fill_dataframe_to_svcg_cycle(df_combined)

    
    df_combined = df_combined.sort_values(['svcg_cycle'], ascending=[True])
    df_combined.reset_index(drop=True, inplace=True)
    return df_combined

def get_data_points(payment_history_IN):
    points = []
    for row in payment_history_IN.iterrows():
        svcg_cycle = str(row[1][0])
        svcg_cycle = float(int(svcg_cycle[0:4]) + float(int(svcg_cycle[4:6]) / 12))  
        x = svcg_cycle # svcg_cycle
        y = row[1][1] # current_upb
        points += [[x,y]]
    return points   
    
loan_ids = get_N_random_loan_ids(50, df)

bad_ids = ['F199Q1000073', 'F199Q1000080', 'F199Q1000087']
good_ids = ['F199Q1000030', 'F199Q1000089', 'F199Q1000088']
good_ids = loan_ids

payment_history_bad = []
for loan_id in bad_ids:
    loan_data_bad = get_loan_data(loan_id, df)
    payment_history_bad_i = get_payment_history(loan_data_bad)
    payment_history_bad += [payment_history_bad_i]

payment_history_good = []
for loan_id in good_ids:
    loan_data_good = get_loan_data(loan_id, df)
    payment_history_good_i = get_payment_history(loan_data_good)
    payment_history_good += [payment_history_good_i]

 
    
payment_history_combined_good = get_combined_payment_history(payment_history_good)
payment_history_combined_bad = get_combined_payment_history(payment_history_bad)

# print payment_history_combined_good
# print payment_history_combined_bad.shape
# print len(svcg_cycles)

points_good = get_data_points(payment_history_combined_good)
points_bad = get_data_points(payment_history_combined_bad)

```

```python
from scipy.interpolate import interp1d

def smooth_line(points_IN):
    x = []
    y = []
    for j in range(0, len(points_IN)):
        x += [points_IN[j][0]]
        y += [points_IN[j][1]]
    
    x_new = np.linspace(min(x), max(x),len(points_IN))
    poly_deg = 4
    coefs = np.polyfit(x, y, poly_deg)
    y_smooth = np.polyval(coefs, x_new)
    
    points = []
    for j in range(0, len(x_new)):
        x_i = x[j]
        if y_smooth[j] < 0 :
            y_i = 0
        else: 
            y_i = y_smooth[j]
        points += [[x_i,y_i]]
    return points


# make line
line_good = {
    'data': [],
    'color': 'rgb(0, 153, 51)',
    'name': 'With Model', 
    'line_width': 2,
    'marker_size': 0.3,
    'mode': 'lines+markers',
}

line_bad = {
    'data': [],
    'color': 'rgb(255, 0, 0)',
    'name': 'Without Model', 
    'line_width': 2,
    'marker_size': 0.3,
    'mode': 'lines+markers',
}

line_good_smoothed = {
    'data': [],
    'color': 'rgb(0, 153, 51)',
    'name': 'With Model (smoothed)',
    'line_width': 0.00,
    'marker_size': 1,
    'mode': 'markers',
}

line_bad_smoothed = {
    'data': [],
    'color': 'rgb(255, 0, 0)',
    'name': 'Without Model (smoothed)',
    'line_width': 0.00,
    'marker_size': 1,
    'mode': 'markers',
}

points_good = get_data_points(payment_history_combined_good)
points_bad = get_data_points(payment_history_combined_bad)

line_good_smoothed['data'] = smooth_line(points_good)
line_bad_smoothed['data'] = smooth_line(points_bad)
line_good['data'] = points_good
line_bad['data'] = points_bad



lines = [line_good, line_bad, line_good_smoothed, line_bad_smoothed]
```

```python
import pandas as pd
import numpy as np
from __future__ import division
import plotly.plotly as py
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot
from plotly.graph_objs import *
from plotly.grid_objs import Grid, Column
from plotly.tools import FigureFactory as FF 
init_notebook_mode(connected=True)
import math


def get_sliders_dict():
    sliders_dict = {
        'active': 0,
        'yanchor': 'top',
        'xanchor': 'left',
        'currentvalue': {
            'font': {'size': 14},
            'prefix': 'YYYY:MM - ',
            'visible': True,
            'xanchor': 'right'
        },
        'transition': {'duration': 300, 'easing': 'cubic-in-out'},
        'pad': {'b': 10, 't': 50},
        'len': 0.9,
        'x': 0.1,
        'y': 0,
        'steps': []
    }
    return sliders_dict

def add_line_animation(figure, line_i, i):
    sliders_dict = get_sliders_dict()
    frames = [] 
    data_dict = []
    start_point = line_i['data'][0]
    x = []
    y = []
    for j in range(0, len(line_i['data'])):
        if line_i['data'][j] == start_point:
            continue
        x += [line_i['data'][j - 1][0], line_i['data'][j][0]]
        y += [line_i['data'][j - 1][1], line_i['data'][j][1]]
        frames.append(
            dict(data= [dict(
                x= x[:],
                y= y[:],
                mode= line_i['mode'],
                marker = dict(
                    size = line_i['marker_size'],
                    color = (line_i['color']),
                    line = dict(
                        width = line_i['line_width'],
                    ), 
                ),
                name= line_i['name']
            )], 
            name= str(j)
        ))
        
        slider_step = {'args': [
            [str(j)],
            {'frame': {'duration': 300, 'redraw': False},
             'mode': 'immediate',
           'transition': {'duration': 300}}
         ],
         'label': str(int(math.floor(line_i['data'][j - 1][0]))) + ":" + str( int(j % 12) + 1 ).zfill(2),
         'method': 'animate'}
        
        sliders_dict['steps'].append(slider_step)

    figure['layout']['sliders'] = [sliders_dict]
    
    # Add frames or add to existing frames   
    if len(figure['frames']) == 0:
        figure['frames'] += frames
    else: 
        for i in range(0, len(figure['frames'])):
            figure['frames'][i]['data'] += frames[i]['data']
    
    # Add start data      
    figure['data'] += [dict(
                        x= [start_point[0], start_point[0]], 
                        y= [start_point[1], start_point[1]],
                        mode= line_i['mode'],
                        marker = dict(
                            size = line_i['marker_size'],
                            color = (line_i['color']),
                            line = dict(
                                width = line_i['line_width'],
                            )
                        ),
                        name= line_i['name']
                      )] 
    return figure



def get_range(data):
    range_buffer_x = 1
    range_buffer_y = 0.1
    x_s = []
    y_s = []
    for point in data:
        x_s += [point[0]]
        y_s += [point[1]]
    x_range = (min(x_s) - range_buffer_x, max(x_s) + range_buffer_x)
    y_range = (min(y_s) - range_buffer_y, max(y_s) + range_buffer_y)
    return (x_range, y_range)


# make figure
figure = {
    'data': [],
    'layout': {},
    'frames': []
}

range_x, range_y = get_range(lines[0]['data'])

for i in range(0,len(lines)): 
    figure = add_line_animation(figure, lines[i], i)
    (x_range, y_range) = get_range(lines[i]['data'])
    range_x = [min(range_x[0], x_range[0]), max(range_x[1], x_range[1])]
    range_y = [min(range_y[0], y_range[0]), max(range_y[1], y_range[1])]



    
    
print range_x
print range_y

axis = dict(ticklen=4,
          mirror=True,
          zeroline=False,
          showline=True,
          autorange=False,
          showgrid=False)

# fill in most of layout
figure['layout']['title'] = "Mortagage Portfolio Analysis - Loan Percentage Paid Over Time"
figure['layout']['xaxis'] = dict(axis, **{'range': range_x, 'title': 'Years'})
figure['layout']['yaxis'] = dict(axis, **{'range': range_y, 'title': 'Loan Percentage Paid'})
figure['layout']['hovermode'] = 'closest'

figure['layout']['updatemenus'] = [
    {
        'buttons': [
            {
                'args': [None, {'frame': {'duration': 100, 'redraw': False},
                         'fromcurrent': True, 'transition': {'duration': 300, 'easing': 'quadratic-in-out'}}],
                'label': 'Play',
                'method': 'animate'
            },
            {
                'args': [[None], {'frame': {'duration': 0, 'redraw': False}, 'mode': 'immediate',
                'transition': {'duration': 0}}],
                'label': 'Pause',
                'method': 'animate'
            }
        ],
        'direction': 'left',
        'pad': {'r': 10, 't': 87},
        'showactive': False,
        'type': 'buttons',
        'x': 0.05,
        'xanchor': 'right',
        'y': 0.05,
        'yanchor': 'top'
    }
]

iplot(figure)
```
